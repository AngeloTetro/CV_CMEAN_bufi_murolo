```python
import os
import pandas as pd
import torch
import numpy as np
import sys
import matplotlib.pyplot as plt
from tqdm import tqdm # Per la barra di progresso
from msclap import CLAP # Stiamo usando la libreria msclap come da te specificato
import traceback # Importa il modulo traceback per stampare i dettagli completi dell'errore
import string # Per la punteggiatura
import nltk 
from nltk.corpus import stopwords # Per le stop words
from nltk.tokenize import word_tokenize # Per dividere il testo in parole

# Scarica le stop words e il tokenizer di NLTK (esegui questo solo una volta)
try:
    nltk.data.find('corpora/stopwords')
except:
    nltk.download('stopwords')
try:
    nltk.data.find('tokenizers/punkt')
except:
    nltk.download('punkt')
# --- AGGIUNTO: Download della risorsa 'punkt_tab' per l'italiano ---
try:
    nltk.data.find('tokenizers/punkt_tab')
except:
    nltk.download('punkt_tab')

# --- 1. Initial Configuration ---
# Determine the device to use (GPU if available, otherwise CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Paths to necessary files and folders.
# *** IMPORTANT: MAKE SURE THESE PATHS ARE CORRECT ON YOUR SYSTEM! ***
PATHS = {
    "csv_file": r"C:\Users\edoar\Downloads\analisi_audio_per_canzone (2).csv",
    # Questo è il percorso della cartella dove si trovano i tuoi file MP3 (es. 000213.mp3)
    "audio_files_folder":"C://Users//edoar//Desktop//PiccoloSubSetAudio", # <--- *** AGGIORNA QUESTO PERCORSO ***
    "clap_weights": r"C://Users//edoar//Downloads//CLAP_weights_2023.pth"
}

# --- 2. Load CLAP Model ---
# Attempt to load the CLAP model. If an error occurs, the program exits.
try:
    clap_model = CLAP(
        version='2023',
        model_fp=PATHS["clap_weights"],
        use_cuda=True if device == "cuda" else False
    )
    print("✅ CLAP model loaded successfully.")
except Exception as e:
    print(f"❌ Error loading CLAP model: {e}")
    print("Make sure the CLAP weights path is correct and the file exists.")
    sys.exit(1) # Exit the script in case of a critical error

# --- 3. Function to Read CSV Data with Specific Structure ---
def leggi_dati_canzoni_da_file_strutturato(percorso_file, riga_iniziale=11, intervallo_righe=9):
    """
    Legge un file riga per riga e estrae il nome del file audio (canzone.mp3)
    e la descrizione dalla riga.

    Args:
        percorso_file (str): Il percorso completo del file CSV.
        riga_iniziale (int): La riga da cui iniziare a cercare il primo dato "nome, descrizione" (1-based).
        intervallo_righe (int): L'intervallo di righe tra un dato e il successivo.

    Returns:
        tuple: Due liste (nomi_file_audio, descrizioni), dove ogni elemento
               corrisponde a una coppia valida trovata. Restituisce ([], []) in caso di errore
               o se non vengono trovati dati.
    """
    nomi_file_audio = []
    descrizioni = []

    if not os.path.exists(percorso_file):
        print(f"Errore: Il file '{percorso_file}' (CSV) non è stato trovato.")
        return [], []

    try:
        with open(percorso_file, 'r', encoding='utf-8') as f:
            for numero_riga, riga_contenuto in enumerate(f, 1):
                if numero_riga >= riga_iniziale and (numero_riga - riga_iniziale) % intervallo_righe == 0:
                    riga_pulita = riga_contenuto.strip()

                    if not riga_pulita:
                        continue

                    parti = riga_pulita.split(',', 1)

                    if len(parti) == 2:
                        nome_audio = parti[0].strip()
                        desc = parti[1].strip()
                        
                        # Rimuovi TUTTI i caratteri di virgoletta (' e ") e i backslash dal nome del file
                        nome_audio = nome_audio.replace('"', '').replace("'", '').replace("\\", '')
                        
                        # Rimuovi eventuali virgolette extra dalla descrizione (solo se all'inizio/fine)
                        if desc.startswith('"') and desc.endswith('"'):
                            desc = desc[1:-1]

                        nomi_file_audio.append(nome_audio)
                        descrizioni.append(desc)
                    else:
                        print(f"Attenzione: Formato riga inatteso alla riga {numero_riga}: '{riga_pulita}'. Non è stata letta come 'nome, descrizione'.")

    except UnicodeDecodeError:
        print(f"Errore: Impossibile leggere il file '{percorso_file}' con l'encoding 'utf-8'.")
        print("Prova a specificare un encoding diverso (es. 'latin-1' o 'cp1252').")
        return [], []
    except Exception as e:
        print(f"Si è verificato un errore inatteso durante la lettura del file '{percorso_file}': {e}")
        return [], []

    return nomi_file_audio, descrizioni

# --- Funzione di pulizia del testo AGGIORNATA ---
def clean_description_text(text):
    """
    Rimuove parole e frasi indesiderate dalla descrizione, normalizza e rimuove stop words.
    """
    if not isinstance(text, str):
        return "" 

    # 1. Converti in minuscolo
    text = text.lower()

    # 2. Rimuovi parole/frasi specifiche che sono "rumore"
    phrases_to_remove = [
        "frammento",
        "descrizione",
        "testo",
        "audio clip",
        "suono di",
        "registrazione di",
        "clip audio",
        "un suono di",
        "una descrizione di",
        "il suono di",
        "la descrizione di",
        "questo è un audio di",
        "questo è un frammento di",
        ".mp3", 
        "brano",
        "canzone",
        "musica",
        "audio" 
    ]
    phrases_to_remove.sort(key=len, reverse=True) 

    for phrase in phrases_to_remove:
        text = text.replace(phrase, " ") 
    
    # 3. Rimuovi punteggiatura
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 4. Rimuovi numeri (se non rilevanti, altrimenti commenta questa riga)
    text = ''.join([i for i in text if not i.isdigit()])

    # 5. Tokenizza il testo e rimuovi le stop words
    words = word_tokenize(text, language='italian') 
    stop_words_italian = set(stopwords.words('italian'))
    
    filtered_words = [word for word in words if word.strip() and word not in stop_words_italian]
    
    # 6. Ricostruisci la stringa
    text = " ".join(filtered_words)
    
    # 7. Rimuovi spazi extra e trimma
    text = " ".join(text.split()).strip()
    
    # 8. Capitalizza la prima lettera della prima parola dopo la pulizia (opzionale)
    if text:
        text = text[0].upper() + text[1:]

    return text

# --- 4. Load Data from CSV ---
print(f"\nReading data from CSV file: {PATHS['csv_file']}")
audio_filenames, descriptions_raw = leggi_dati_canzoni_da_file_strutturato(PATHS["csv_file"])

if not audio_filenames:
    print("❌ Error: No valid audio filenames and descriptions extracted from CSV.")
    sys.exit(1)

# Applica la pulizia alle descrizioni
descriptions_cleaned = [clean_description_text(desc) for desc in descriptions_raw]

print(f"✅ Extracted {len(audio_filenames)} audio filenames and {len(descriptions_cleaned)} cleaned descriptions from CSV.")

# --- 5. Extract Audio and Text Embeddings for each pair ---
audio_embeddings_list = []
text_embeddings_list = []
processed_pairs_info = [] 

print("\nExtracting embeddings and computing similarities:")
for i in tqdm(range(len(audio_filenames)), desc="Processing Pairs"):
    filename = audio_filenames[i]
    description_for_embedding = descriptions_cleaned[i] 
    description_original = descriptions_raw[i] 
    
    # Define a maximum character length for the description to prevent potential tokenizer issues
    MAX_TEXT_CHAR_LENGTH = 512 
    
    # Truncate the description if it's too long (applica alla descrizione pulita)
    if isinstance(description_for_embedding, str) and len(description_for_embedding) > MAX_TEXT_CHAR_LENGTH:
        truncated_description = description_for_embedding[:MAX_TEXT_CHAR_LENGTH]
        if not truncated_description.endswith('...'): 
             truncated_description += "..."
    else:
        truncated_description = description_for_embedding

    # Construct the full path to the audio file
    full_audio_path = os.path.join(PATHS["audio_files_folder"], filename)

    if not os.path.exists(full_audio_path):
        print(f"❌ Skipped '{filename}': Audio file not found at '{full_audio_path}'.")
        continue 
    
    # Salta se la descrizione è diventata vuota dopo la pulizia (potrebbe accadere se conteneva solo parole da rimuovere)
    if not truncated_description.strip():
        print(f"⚠️ Skipped '{filename}': Description became empty after cleaning. Original: '{description_original}'.")
        continue

    try:
        # Get audio embedding
        audio_emb = clap_model.get_audio_embeddings([full_audio_path], resample=True)
        
        # Get text embedding using the (potentially) truncated and cleaned description
        text_emb = clap_model.get_text_embeddings([truncated_description]) 

        # Add to lists
        audio_embeddings_list.append(audio_emb)
        text_embeddings_list.append(text_emb)
        processed_pairs_info.append({'filename': filename, 'description': description_for_embedding})

    except IndexError as e: 
        print(f"❌ ERROR (IndexError) processing '{filename}' (description: '{truncated_description[:50]}...'): {e}")
        print("    This likely means the text description was problematic (e.g., too long after truncation, or invalid characters). Skipping this pair.")
        print("    Full Traceback:")
        traceback.print_exc(file=sys.stdout) 
        continue 
    except Exception as e: 
        print(f"❌ GENERAL ERROR processing '{filename}': {e}")
        print("    Full Traceback:")
        traceback.print_exc(file=sys.stdout)
        continue

if not audio_embeddings_list:
    print("\n❌ Error: No audio-text pairs were successfully processed for embedding extraction. Check audio file paths and formats.")
    sys.exit(1)

# Concatenate embeddings into single tensors
audio_embeddings_tensor = torch.vstack([e.squeeze(0) for e in audio_embeddings_list])
text_embeddings_tensor = torch.vstack([e.squeeze(0) for e in text_embeddings_list])

print(f"✅ Successfully extracted embeddings for {len(audio_embeddings_tensor)} pairs.")

# --- 6. Calculate Cosine Similarities ---
audio_embeddings_norm = torch.nn.functional.normalize(audio_embeddings_tensor, p=2, dim=-1)
text_embeddings_norm = torch.nn.functional.normalize(text_embeddings_tensor, p=2, dim=-1)

similarities = (audio_embeddings_norm * text_embeddings_norm).sum(dim=-1)

similarities_np = similarities.cpu().detach().numpy()

print(f"✅ Calcolate {len(similarities_np)} similarità audio-testo.")

# --- 7. Visualizzazione Distribuzione Similarità ---
plt.figure(figsize=(10, 6))
plt.hist(similarities_np, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribuzione Similarità Audio-Descrizione (CLAP)")
plt.xlabel("Similarità Coseno")
plt.ylabel("Frequenza")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 8. Statistiche ---
print("\n=== Statistiche Similarità Audio-Descrizione (CLAP) ===")
print(f"Numero di coppie elaborate: {len(similarities_np)}")
print(f"Media: {np.mean(similarities_np):.4f}")
print(f"Mediana: {np.median(similarities_np):.4f}")
print(f"Deviazione standard: {np.std(similarities_np):.4f}")
print(f"Minima: {np.min(similarities_np):.4f}")
print(f"Massima: {np.max(similarities_np):.4f}")

# --- 9. Visualizza Esempi Migliori e Peggiori ---
def print_similarity_examples(indices, title, data_info, similarities_scores, num_examples=5):
    """
    Stampa i dettagli degli esempi con le similarità più alte o più basse.
    data_info deve essere la lista `processed_pairs_info` che contiene solo le tracce valide.
    """
    print(f"\n=== {title} ===")
    for i, idx in enumerate(indices):
        if i >= num_examples:
            break
        filename = data_info[idx]['filename']
        description = data_info[idx]['description'] 
        sim_score = similarities_scores[idx]
        print(f"File: {filename}, Descrizione: '{description[:100]}...', Similarità: {sim_score:.4f}")

# Ordina gli indici per similarità (dal più basso al più alto)
sorted_indices = np.argsort(similarities_np)

# Visualizza i 5 peggiori match (similarità più basse)
print_similarity_examples(sorted_indices[:5], "🔻 Peggiori match Audio-Descrizione", processed_pairs_info, similarities_np)

# Visualizza i 5 migliori match (similarità più alte)
print_similarity_examples(sorted_indices[-5:][::-1], "🔺 Migliori match Audio-Descrizione", processed_pairs_info, similarities_np)
```
